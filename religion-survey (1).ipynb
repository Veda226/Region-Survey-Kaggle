```python
religion_patterns = [
    (df['What is your present religion, if any?'].str.contains('Bap', case=False, regex=False), 'Baptist'),
   (df['What is your present religion, if any?'].str.contains('Bath', case=False, regex=False), 'Baptist'),
    (df['What is your present religion, if any?'].str.contains('Babtist', case=False, regex=False), 'Baptist'),
     (df['What is your present religion, if any?'].str.contains('Athe', case=False, regex=False), 'Atheist'),
   (df['What is your present religion, if any?'].str.contains('Christ', case=False, regex=False), 'Christian'),
    (df['What is your present religion, if any?'].str.contains('Don', case=False, regex=False), 'None'),
    (df['What is your present religion, if any?'].str.contains('I ', case=False, regex=False), 'None'),
   (df['What is your present religion, if any?'].str.contains('Luthe', case=False, regex=False), 'Lutheran'),
    (df['What is your present religion, if any?'].str.contains('Jehovah', case=False, regex=False), 'Jehovah'),
    (df['What is your present religion, if any?'].str.contains('Meth', case=False, regex=False), 'Methodist'),
   (df['What is your present religion, if any?'].str.contains('My ', case=False, regex=False), 'None'),
    (df['What is your present religion, if any?'].str.contains('NONE', case=False, regex=False), 'None'),
    (df['What is your present religion, if any?'].str.contains('No ', case=False, regex=False), 'None'),
   (df['What is your present religion, if any?'].str.contains('spirit ', case=False, regex=False), 'Spritual'),
    (df['What is your present religion, if any?'].str.contains('A ', case=False, regex=False), 'None'),   
        (df['What is your present religion, if any?'].str.contains('Catholic', case=False, regex=False), 'Catholic'),
   (df['What is your present religion, if any?'].str.contains('do ', case=False, regex=False), 'None'),
          (df['What is your present religion, if any?'].str.contains('Pentecost', case=False, regex=False), 'Pentecostal'),
    (df['What is your present religion, if any?'].str.contains('Pagan', case=False, regex=False), 'Pagan'),
       (df['What is your present religion, if any?'].str.contains('A ', case=False, regex=False), 'None') 
]
```

```python
religion_criteria = ""
religion_values = ""
```

```python
religion_criteria, religion_values = zip(*religion_patterns)
df['religion_Group_1'] = np.select(religion_criteria, religion_values, df['What is your present religion, if any?'])
```

```python
pd.crosstab(df['religion_Group_1'],df['What is your age?'])
```

```python
df_earn = df['How much total combined money did all members of your HOUSEHOLD earn last year?'].str.split("to", n = 1, expand = True)
```

```python
df_earn.head()
```

```python
df = pd.concat([df, df_earn], axis=1)
```

```python
df.head()
```

```python
print(df.groupby([1]).mean())
```

```python
print(df.groupby([0]).mean())
```

```python
earning_patterns = [
    (df[0].str.contains('and up', case=False, regex=False), '$200,000'),
   (df[0].str.contains('Prefer not', case=False, regex=False), ''),
    (df[0].str.contains('Response', case=False, regex=False), '')
]
```

```python
earning_criteria, earning_values = zip(*earning_patterns)
df['min earn'] = np.select(earning_criteria, earning_values, df[0])
```

```python
df['min earn']=df['min earn'].replace("", "0")
```

```python
earning_patterns_max = [
    (df[1].str.contains(' answer', case=False, regex=False, na=False), '')
]
```

```python
earning_criteria_max, earning_values_max = zip(*earning_patterns_max)
df['max earn'] = np.select(earning_criteria_max, earning_values_max, df[1])
```

```python
print(df.groupby(['max earn']).mean())
```

```python
df['min earn'] = df['min earn'].str.replace('$', '')
df['max earn'] = df['max earn'].str.replace('$', '')
```

```python
print(df.groupby(['min earn']).mean())
```

```python
df['min earn'] = df['min earn'].fillna(0)
```

```python
df['min earn'] = df['min earn'].str.strip()
```

```python
df['min earn'] = df['min earn'].str.replace(',', '')
df['min earn'] = df['min earn'].astype(str).astype(int)
```

```python
print(df.groupby(['min earn']).mean())
```

```python
df['min earn bin'] = pd.qcut(df['min earn'], q=10, precision=0, duplicates='drop')
```

```python
pd.crosstab(df['religion_Group_1'],df['min earn bin'])
```
